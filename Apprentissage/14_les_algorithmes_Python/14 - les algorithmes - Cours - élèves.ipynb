<h3> Le tri par selection </h3>
    
Le tri par sélection est un algorithme de tri simple qui trie une liste en trouvant répétitivement le minimum de la liste non triée et en le plaçant au début. Voici comment vous pouvez le construire étape par étape en Python.

<b> Étape 3: Combiner les Étapes pour le Tri par Sélection <b>

- Boucle Principale : La boucle for parcourt chaque élément de la liste. À chaque itération, l'élément à l'index i est considéré comme le premier élément de la liste non triée.
- Trouver l'Élément Minimum : Pour chaque itération de la boucle, la fonction trouver_minimum est appelée pour trouver l'index de l'élément le plus petit dans la liste non triée (de i à la fin de la liste).
- Échanger les Éléments : Une fois l'index du minimum trouvé, cet élément est échangé avec l'élément à l'index i (le début de la liste non triée) à l'aide de la fonction echanger.

```python
liste_exemple = [4,1,2,3]
```

<b> Étape 2: Parcourir la Liste pour Appliquer les Échanges </b>

La seconde étape consiste à appliquer cette comparaison et cet échange à travers toute la liste, pour une seule itération.

<b> Étape 3: Répéter le Processus </b>

Le tri à bulles nécessite de répéter le processus de comparaison et d'échange plusieurs fois jusqu'à ce que la liste soit entièrement triée.
