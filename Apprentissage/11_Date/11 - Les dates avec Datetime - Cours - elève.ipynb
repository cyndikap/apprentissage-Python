<h1> Introduction </h1>

Le module datetime en Python est une bibliothèque conçue pour manipuler les dates et les heures. Dans cette bibliothèque se trouvent plusieurs classes principales qui facilitent la gestion du temps dans les applications Python : 
- date : Gère les dates au format année, mois, jour.
- time : Représente un moment de la journée, indépendant d'une date spécifique.
- datetime : Combine les informations de date et de temps.
- timedelta : Représente une durée, la différence entre deux dates ou temps.
- timezone : Gère les informations de fuseau horaire pour les objets datetime.

<h1> La classe Time </h1>

<b> créer et à utiliser des objets time </b>

Un objet time est utilisé pour représenter un moment de la journée, sans lien avec une date spécifique. Les paramètres heure, minute, seconde, et microseconde sont optionnels et définissent le moment précis.

Syntaxe : datetime.time(heure, minute, seconde, microseconde)

<b> Gestion des fuseaux horaires </b>

La gestion des fuseaux horaires est importante pour les applications nécessitant de travailler avec des utilisateurs ou des événements dans plusieurs régions géographiques. datetime.timezone permet une gestion précise des fuseaux horaires, y compris le calcul des décalages horaires
